```python
import json, re
from collections import Counter
from datasets import load_dataset

# Load your final IV bigrams from Step 1
with open("/content/combined_tts_iv_clean.json", encoding="utf-8") as f:
    iv_data = json.load(f)

IV_BIGRAMS = set(iv_data["iv_bigrams"])
IV_WORDS   = set(iv_data["iv_words"])

print(f"IV bigrams loaded : {len(IV_BIGRAMS):,}")
print(f"IV words loaded   : {len(IV_WORDS):,}")

# Strict Devanagari filter — same as Step 1, no digits, no punctuation
PURE_DEVA = re.compile(r'^[\u0900-\u0963\u0972-\u097F]+$')
STRIP_CHARS = '।,\'\"''""\u200b\u200c\u200d()[]{}!?:;—–-#@$%^&*_~`|\\/<>…'

def clean_token(token: str) -> str | None:
    t = token.strip(STRIP_CHARS)
    if len(t) >= 2 and PURE_DEVA.match(t):
        return t
    return None
```

```python
DS1 = "Titung/cc100-nepali-cleaned"

ds1 = load_dataset(DS1, split="train", streaming=True, trust_remote_code=True)

# peek at structure
sample = next(iter(ds1))
print("Columns :", list(sample.keys()))
print("Sample  :", sample)
```

```python
# ── ADJUST if field name differs ──
TEXT_FIELD_DS1 = "text"
# ─────────────────────────────────

corpus_words_ds1 = Counter()
ds1_total_tokens = 0
ds1_rows = 0

for ex in load_dataset(DS1, split="train", streaming=True, trust_remote_code=True):
    for tok in ex[TEXT_FIELD_DS1].split():
        ds1_total_tokens += 1
        w = clean_token(tok)
        if w:
            corpus_words_ds1[w] += 1
    ds1_rows += 1
    if ds1_rows % 50_000 == 0:
        print(f"  {ds1_rows:,} rows | {ds1_total_tokens:,} tokens | {len(corpus_words_ds1):,} unique clean words")

print(f"\nDS1 done — {ds1_rows:,} rows | {len(corpus_words_ds1):,} unique clean Devanagari words")
```

```python
DS2 = "Titung/nepali-brihat-sabdakosh"

ds2 = load_dataset(DS2, split="train", trust_remote_code=True)
print("Columns :", ds2.column_names)
print("Sample  :", ds2[0])
```

```python
# ── ADJUST if field name differs ──
# Dictionary datasets often have 'word', 'definition', 'meaning' etc.
# Collect from ALL text fields to maximise vocabulary coverage
TEXT_FIELDS_DS2 = ["word", "definition"]   # adjust after inspecting columns
# ─────────────────────────────────────────

corpus_words_ds2 = Counter()

for ex in ds2:
    for field in TEXT_FIELDS_DS2:
        text = ex.get(field, "") or ""
        for tok in text.split():
            w = clean_token(tok)
            if w:
                corpus_words_ds2[w] += 1

print(f"DS2 done — {len(ds2):,} rows | {len(corpus_words_ds2):,} unique clean Devanagari words")
print(f"Sample words: {list(corpus_words_ds2.keys())[:20]}")
```

```python
DS3 = "Titung/nepali-asr"

ds3 = load_dataset(DS3, split="train", trust_remote_code=True)
print("Columns :", ds3.column_names)
print("Sample  :", ds3[0])
```

```python
# ASR dataset — use only the 'transcription' field (Nepali text)
# Audio columns (audio, audio_duration_s, utterance_id, speaker_id) are ignored
TEXT_FIELD_DS3 = "transcription"

corpus_words_ds3 = Counter()

for ex in ds3:
    text = ex.get(TEXT_FIELD_DS3, "") or ""
    for tok in text.split():
        w = clean_token(tok)
        if w:
            corpus_words_ds3[w] += 1

print(f"DS3 done — {len(ds3):,} rows | {len(corpus_words_ds3):,} unique clean Devanagari words")
print(f"Sample words: {list(corpus_words_ds3.keys())[:20]}")
```

```python
import os, json

# Union of all four counters
corpus_all = Counter()
for c in [corpus_words_ds1, corpus_words_ds2, corpus_words_ds3]:
    corpus_all.update(c)

# Remove words already in IV set
corpus_novel = Counter({w: c for w, c in corpus_all.items() if w not in IV_WORDS})

# Filter low-frequency noise — words appearing only once are likely OCR errors,
# typos, or web garbage. Minimum freq=2 cuts the file size dramatically.
MIN_FREQ = 2
corpus_novel_filtered = Counter({w: c for w, c in corpus_novel.items() if c >= MIN_FREQ})

print("=" * 55)
print("STEP 2 — CORPUS SUMMARY")
print("=" * 55)
print(f"  cc100-nepali-cleaned    : {len(corpus_words_ds1):>10,} unique words")
print(f"  nepali-brihat-sabdakosh : {len(corpus_words_ds2):>10,} unique words")
print(f"  nepali-asr (transcripts): {len(corpus_words_ds3):>10,} unique words")
print(f"  {'─'*42}")
print(f"  UNION total             : {len(corpus_all):>10,} unique words")
print(f"  Already in IV (skip)    : {len(corpus_all) - len(corpus_novel):>10,} words")
print(f"  Novel (raw)             : {len(corpus_novel):>10,} words")
print(f"  Novel (freq >= {MIN_FREQ})      : {len(corpus_novel_filtered):>10,} words  ← saved")

# Save
os.makedirs("step2_results", exist_ok=True)

payload = {
    "description": "Step 2 corpus — union of 4 sources, cleaned Devanagari, IV words removed, freq >= 2",
    "sources": [DS1, DS2, DS3],
    "min_freq": MIN_FREQ,
    "n_total_unique_words":  len(corpus_all),
    "n_iv_words_removed":    len(corpus_all) - len(corpus_novel),
    "n_novel_words_raw":     len(corpus_novel),
    "n_novel_words_saved":   len(corpus_novel_filtered),
    # list of {"word": w, "freq": f} sorted by freq descending
    "novel_words": [
        {"word": w, "freq": f}
        for w, f in corpus_novel_filtered.most_common()
    ],
}

out_path = "step2_results/corpus_novel_words.json"
with open(out_path, "w", encoding="utf-8") as f:
    json.dump(payload, f, ensure_ascii=False, indent=2)

print(f"\n✓ Saved → {out_path}")
print(f"  Step 2 complete. Feed novel_words into Step 3 OOV detection.")
```
